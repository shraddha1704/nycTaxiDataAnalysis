```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
data = pd.read_csv('citibikes_clean_wo_weather.csv')
#print data.values
dist =  data[['gender','trip_duration']]
```

```python
distVals = dist.values
mean = np.mean(dist[['trip_duration']])
stdDev = np.std(dist[['trip_duration']])
T = abs(dist[['trip_duration']] - mean) < .75 * stdDev
#T = abs(dist[['trip_duration']])  < 2000
#print T
data_wo_outlier = distVals[np.array(T).flatten()]
```

```python
male=[]
female=[]
for i in data_wo_outlier:
    if(i[0]==1):
        male.append(i[1])
    elif(i[0]==2):
        female.append(i[1])
print len(male)
```

```python
sns.distplot(male)
sns.distplot(female)
```
