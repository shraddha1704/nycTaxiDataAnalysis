```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
fileData = pd.read_csv('trips_clean_500000.csv')
data = fileData[['start_time']]
data = np.array(data).flatten()
dic = collections.defaultdict(int)
#print data
```

```python
array = ['Thursday', 'Friday', 'Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday']
dicDay = collections.defaultdict(int)
count = 0
for val in array:
    dicDay[val] = count
    count += 1
#print dicDay
```

```python
first = parse(data[0])
#print first.day
for day in data:
    day = parse(day)
    dic[(day.day - first.day)%7] += 1

#print dic

    
```

```python
result = []
days = ['Monday', 'Tuesday', 'Wednesday','Thursday', 'Friday', 'Saturday', 'Sunday']
for val in range(7):
    result.append(dic[(val+4)%7])
print result
print days
plt.plot(result, label = "Bike Usage")
plt.xticks([i for i in range(7)], days, rotation='vertical')

plt.ylabel('CitiBike Count')
plt.xlabel('Day')
plt.legend(loc='upper left')
plt.show()
```
