```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn import linear_model
from dateutil.parser import parse
from sklearn.metrics import mean_squared_error
```

```python
def MSE(Y_actual, Y_predicted):
    sum = 0
    for i in range (len(Y_actual)):
        sum+=abs(Y_predicted[i] - Y_actual[i])
    return sum/len(Y_predicted)
def MAPE(Y_actual, Y_predicted):
    return (100.0/len(Y_actual))* sum(abs((Y_predicted-Y_actual)/(Y_actual)))
```

```python
data = pd.read_csv('CitibikeTripsWeathers_100000.csv')
```

```python
bucket = 1
hrCount = data[['start_time']]
hrCount = np.array(hrCount).flatten()
```

```python
hr1 = parse(min(hrCount))
array = [0 for i in range(20000)]
```

```python
hour = hr1
for t in hrCount:
    t = parse(t)
    #print t
    time = t - hour
    diff = ((time.days*24) + (time.seconds)//3600) / bucket
    array[diff] += 1
#print array
    
```

```python
result = []
#print data
#data = data.values
predictTrue= 10
for i in range(150):
    result.append(array[predictTrue+i-1])
Y=[]
for i in range(10,160):
    Y.append(array[i])
print len(Y)
print len(result)
```
