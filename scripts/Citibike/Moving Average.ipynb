```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn import linear_model
from dateutil.parser import parse
```

```python
data = pd.read_csv('CitibikeTripsWeathers_100000.csv')
```

```python
bucket = 1
hrCount = data[['start_time']]
hrCount = np.array(hrCount).flatten()
```

```python
hr1 = parse(min(hrCount))
array = [0 for i in range(20000)]
```

```python
hour = hr1
for t in hrCount:
    t = parse(t)
    #print t
    time = t - hour
    diff = ((time.days*24) + (time.seconds)//3600) / bucket
    array[diff] += 1
```

```python
def avg(data,index, window):
    x=[]
    for i in range(window):
        x.append(data[index-i-1])
    return np.mean(x)
```

```python
result = []
#print dat
#data = data.values
predictTrue= 200
for i in range(150):
    result.append(avg(array,predictTrue+i,6))
Y=[]
for i in range(200,350):
    Y.append(array[i])
print len(Y)
print len(result)
```

```python
#T = data[577:]
print
plt.plot(Y, label = "actual")
plt.plot(result, label="predicted")
sum = 0.0
for i in range(len(result)):
    sum += abs(result[i]-Y[i])
print "average error:  ",str(sum/len(result))
plt.xlabel('Hourly Data')
plt.ylabel('Hourly Usage')
#print plt
plt.legend()
plt.show()
```

```python
def SSE(Y_actual, Y_predicted):
    return sum((Y_predicted - Y_actual)**2)/len(Y_predicted)
def MAPE(Y_actual, Y_predicted):
    return (100.0/len(Y_actual))* sum(abs((Y_predicted-Y_actual)/(Y_actual)))
```
