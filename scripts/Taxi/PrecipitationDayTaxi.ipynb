```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
fileData = open('taxi_raw_clean_10000000.csv','r')

day1dic = collections.defaultdict(int)
day2dic = collections.defaultdict(int)
#days = collections.defaultdict(list)
```

```python
bucket = 1
day1 = '2014-04-30'
day2 = '2014-05-07'
data = fileData.readline()
#data = data.split(',')
#print data[10]
lines = fileData.readlines()
for val in lines:
    val = val.split(',')
    #print val
    date = val[0]
    dateCopy = date.split(' ')
    eachDate = dateCopy[0]
    if eachDate == day1:
        parsedDate = parse(date)
        day1dic[parsedDate.hour] += 1
        
    if eachDate == day2:
        parsedDate = parse(date)
        day2dic[parsedDate.hour] += 1
            
print day1dic
print day2dic
```

```python
plt.plot(day1dic.values(), label = 'Precipitation')
plt.plot(day2dic.values(), label = 'No Precipitation')
plt.xticks([i for i in range(24)], rotation='vertical')

plt.ylabel('Number of Taxis')
plt.xlabel('Hours')
plt.legend(loc = 'upper left')
plt.show()
#plt.savefig('Weather.png')
```
